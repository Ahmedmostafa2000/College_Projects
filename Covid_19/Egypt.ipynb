```python
from matplotlib import pyplot
def determine_country(country):
    data_country = pd.DataFrame(data.loc[[country]])
    data_country = data_country.set_index("Last Update")
    data_country.to_csv("{0}_1603.csv".format(country))
    %matplotlib inline
    data_country.Confirmed.plot(kind="line")
    return data_country
```
